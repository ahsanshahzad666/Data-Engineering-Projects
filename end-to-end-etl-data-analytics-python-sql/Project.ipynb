```python
import pandas as pd
ds = pd.read_csv('orders.csv', na_values=['unknown', 'Not Available'])
ds['Ship Mode'].unique()
```

```python
# ds = ds.rename(columns={'Order Id':'order_id'})
ds
ds.head(10)
```

```python
ds.columns = ds.columns.str.lower()
ds.columns = ds.columns.str.replace(' ', '_')
# ds
# # ds.head(20)
```

```python
ds['discount_amount'] = ds['list_price']*ds['discount_percent']*.01
```

```python
ds['sale_price'] = ds['list_price'] - ds['discount_amount']
# ds
# ds.head(20)
```

```python
ds['profit'] = ds['sale_price'] - ds['cost_price']
# ds.head(20)
```

```python
type(ds['order_date'])
ds.info()
```

```python
ds['order_date'].dtype
```

```python
ds['order_date'].dtype
```

```python
ds['order_date'] = pd.to_datetime(ds['order_date'], format = "%Y-%m-%d")
# ds
```

```python
ds = ds.drop(columns=['cost_price', 'discount_percent', 'list_price'])
```
